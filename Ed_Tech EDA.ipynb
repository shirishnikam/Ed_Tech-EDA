```python
# Dropping the columns which is not needed
data.drop(["url","avg_rating_recent","avg_rating","discount_price__currency",      
                 "discount_price__price_string","price_detail__currency",
                 "price_detail__price_string","created","published_time"],axis=1,inplace=True)
```

```python
data.fillna(data["discount_price__amount"].mean(),inplace=True)
```

```python
data.isnull().sum()
```

```python
## converting the date-time into month & year
data["created_year"] = pd.to_datetime(data["created"]).dt.year
data["created_month"] = pd.to_datetime(data["created"]).dt.month
data["publish_year"] = pd.to_datetime(data["published_time"]).dt.year
data["publish_month"] = pd.to_datetime(data["published_time"]).dt.month
```

```python
# converting the avg_rating float variable to int for improving analysis

data["Rating_recent"] = np.where(data["rating"]>=4.5,5,
                        np.where(data["rating"]>=4,4,
                        np.where(data["rating"]>=3.5,4,
                        np.where(data["rating"]>=3,3,
                        np.where(data["rating"]>=2.5,3,
                        np.where(data["rating"]>=2,2,
                        np.where(data["rating"]>=1.5,2,
                        np.where(data["rating"]>=1,1,
                        np.where(data["rating"]>=0.5,1,0)))))))))
```
